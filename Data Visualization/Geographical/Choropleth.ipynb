```python
import chart_studio.plotly as py
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import plotly.graph_objs as go
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected=True)
```

```python
data = dict(type='choropleth',
           locations=['AZ','CA','NY'],
           locationmode= 'USA-states',
           colorscale='Jet',
           text=['text 1','text 2','text 3'],
           z = [1.0,2.0,3.0],
           colorbar={'title':'Colorbar Title Goes Here'})
```

```python
layout = dict({'scope':'usa'})
```

```python
choromap = go.Figure(data = [data],layout = layout,skip_invalid = True)
```

```python
iplot(choromap)
```

```python
df = pd.read_csv('2011_US_AGRI_Exports.csv')
```

```python
df.head()
```

```python
data1 = dict(type='choropleth',
            colorscale = 'YIOrRd',
            locations = df['code'],
            z = df['total exports'],
            locationmode = 'USA-states',
            text = df['text'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width = 2)),
            colorbar = {'title':"Millions USD"}
            ) 
```

```python
layout1 = dict(title = '2011 US Agriculture Exports by State',
              geo = dict(scope='usa',
                         showlakes = True,
                         lakecolor = 'rgb(85,173,240)')
             )
```

```python
choromap1 = go.Figure(data = [data1],layout = layout1,skip_invalid = True)
```

```python
iplot(choromap)
```

```python
df = pd.read_csv('2014_World_GDP.csv')
df.head()
```

```python
data2 = dict(
        type = 'choropleth',
        locations = df['CODE'],
        z = df['GDP (BILLIONS)'],
        text = df['COUNTRY'],
        colorbar = {'title' : 'GDP Billions US'},
      ) 
```

```python
layout2 = dict(
    title = '2014 Global GDP',
    geo = dict(
        showframe = False,
        projection = {'type':'natural earth'}
    )
)
```

```python
choromap2 = go.Figure(data = [data2],layout = layout2,skip_invalid=True)
```

```python
iplot(choromap2)
```
